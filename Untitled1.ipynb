```python
# Step 1: Mount Drive and import libraries
from google.colab import drive
import geopandas
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt

drive.mount('/content/drive', force_remount=True)

# Step 2: Load both datasets
vector_url = "https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip"
world = geopandas.read_file(vector_url)

raster_path = '/content/drive/My Drive/T30UWF_20250103T112451_B04_10m.jp2'

# Step 3: Reproject and Plot
with rasterio.open(raster_path) as dataset:
    # Reproject the vector data to match the raster's CRS
    world_projected = world.to_crs(dataset.crs)

    # --- NEW STEP: Clip the projected vector data to the raster's extent ---
    # We get the raster's bounding box to create a filtering geometry
    left, bottom, right, top = dataset.bounds

    # The .cx indexer is a fast way to select data within a bounding box.
    world_clipped = world_projected.cx[left:right, bottom:top]

    # Create the plot
    fig, ax = plt.subplots(figsize=(10, 10))

    # First, plot the raster image
    show(dataset, ax=ax, cmap='gray', title="Clipped Vector Overlay on Raster")

    # Next, plot the CLIPPED vector data on the SAME axis
    world_clipped.plot(ax=ax, facecolor='none', edgecolor='yellow', linewidth=2)

    plt.show()

print("\nModule 1 is now complete!")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Convert data to float to allow for decimal and negative values
b4 = b4.astype(float)
b8 = b8.astype(float)

# 2. Calculate NDVI
# We suppress a potential "divide by zero" warning for pixels where (b8 + b4) is 0
np.seterr(divide='ignore', invalid='ignore')
ndvi = (b8 - b4) / (b8 + b4)

# 3. Visualize the NDVI result
#plt.figure(figsize=(10, 10))
#plt.imshow(ndvi, cmap='RdYlGn') # 'RdYlGn' is a good colormap for NDVI
#plt.colorbar(label='NDVI')
#plt.title('NDVI Result (Before Cloud Masking)')


print("\nStep 2 complete. You've calculated and visualized the NDVI.")
print("Notice there may be some strange values or noise caused by clouds.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the SCL values that we want to mask
# These values correspond to clouds, cloud shadows, snow, etc.
# Values to mask:
# 1 = Saturated or defective
# 3 = Cloud shadows
# 8 = Cloud medium probability
# 9 = Cloud high probability
# 10 = Thin cirrus
# 11 = Snow or ice
bad_scl_values = [1, 3, 8, 9, 10, 11]

# 2. Create a boolean mask from the SCL data
# The result is an array where 'True' indicates a bad pixel.
cloud_mask_20m = np.isin(scl, bad_scl_values)

# 3. Upscale the mask from 20m to 10m resolution to match our NDVI
# We repeat each pixel 2x horizontally and 2x vertically.
# This is a simple way to perform "nearest neighbor" resizing.
cloud_mask_10m = np.repeat(cloud_mask_20m, 2, axis=0).repeat(2, axis=1)

# 4. Visualize the final mask
#plt.figure(figsize=(10, 10))
#plt.imshow(cloud_mask_10m, cmap='gray')
#plt.title('Final Cloud & Shadow Mask (10m)')


print("\nStep 3 complete. The white areas in the plot are the pixels we will remove.")
```

```python
import numpy as np
import numpy.ma as ma

# --- Assume 'ndvi' and 'cloud_mask_10m' are already loaded ---
# For demonstration purposes, let's create dummy data.
# In your actual use case, you would have your real data here.
#np.random.seed(0)
#ndvi = np.random.rand(100, 100) * 2 - 1  # NDVI values from -1 to 1
#cloud_mask_10m = np.random.choice([True, False], size=(100, 100), p=[0.2, 0.8]) # 20% cloud cover

# 1. Create a masked array where the cloud mask is True.
#    This is memory-efficient as it doesn't copy the ndvi data.
ndvi_masked = ma.masked_where(cloud_mask_10m, ndvi)

print("--- Verifying Cloud Mask Application ---")

# --- Verification Method 1: Count the Masked Pixels ---
# This is the most direct way to confirm the mask was applied.
print("\n1. Verifying by counting pixels:")

# Count the number of True values in the original boolean cloud mask.
num_cloud_pixels = np.sum(cloud_mask_10m)

# Count the number of elements in the mask of our new masked array.
# The .mask attribute returns a boolean array of the same shape.
num_masked_pixels = ndvi_masked.mask.sum()

print(f"   - Number of 'True' pixels in cloud_mask_10m: {num_cloud_pixels}")
print(f"   - Number of masked pixels in ndvi_masked:     {num_masked_pixels}")

# Assert that the numbers match. If they don't, this will raise an error.
assert num_cloud_pixels == num_masked_pixels
print("\n   ✅ SUCCESS: The number of masked pixels matches the cloud mask.")


# --- Verification Method 2: Compare Summary Statistics ---
# Mathematical operations on a masked array automatically ignore masked values.
# Therefore, the mean of the masked array should be different from the original.
print("\n2. Verifying by comparing statistics:")

# Calculate the mean of the original array (includes cloud values)
original_mean = np.mean(ndvi)

# Calculate the mean of the masked array (ignores cloud values)
masked_mean = ndvi_masked.mean() # or np.mean(ndvi_masked)

print(f"   - Mean NDVI of original image (with clouds): {original_mean:.4f}")
print(f"   - Mean NDVI of masked image (without clouds):  {masked_mean:.4f}")

if not np.isclose(original_mean, masked_mean):
    print("\n   ✅ SUCCESS: The mean value changed, confirming the mask is active.")
else:
    print("\n   ⚠️ WARNING: The mean value did not change. Check your data.")

print("\n--- Verification Complete ---")
```

```python
import numpy as np
import numpy.ma as ma
import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling

with rasterio.open(b4_path) as src:
    src_crs = src.crs
    src_transform = src.transform
    width = src.width
    height = src.height
    profile = src.profile
    bounds = src.bounds


    print("📌 Raster Metadata:")
    print(f"CRS: {src_crs}")
    print(f"Transform (Affine):\n{src_transform}")
    print(f"Width: {width}")
    print(f"Height: {height}")
    print(f"Profile:\n{profile}")

    bounds = src.bounds  # Much more accurate


    dst_crs = 'EPSG:27700'
    dst_transform, dst_width, dst_height = calculate_default_transform(
    src_crs, dst_crs, width, height, *bounds
)

    ndvi_filled = ma.filled(ndvi_masked, fill_value=-9999).astype(np.float32)

# Create an empty destination array for the reprojected data
dst_ndvi = np.full((dst_height, dst_width), -9999, dtype=np.float32)

reproject(
    source=ndvi_filled,
    destination=dst_ndvi,
    src_transform=src_transform,
    src_crs=src_crs,
    dst_transform=dst_transform,
    dst_crs=dst_crs,
    resampling=Resampling.bilinear  # you can also use Resampling.nearest
)


# Update the profile for the new file
profile.update({
    'crs': dst_crs,
    'transform': dst_transform,
    'width': dst_width,
    'height': dst_height,
    'nodata': -9999,
    'dtype': 'float32',
    'driver':'GTiff',

})

# Save the output file
folder_path = '/content/drive/My Drive/'
output_path = f'{folder_path}ndvi_reprojected_epsg27700.tif'
with rasterio.open(output_path, 'w', **profile) as dst:
    dst.write(dst_ndvi, 1)

print(f"✅ Reprojected NDVI saved at: {output_path}")

with rasterio.open(output_path) as src:
    print(f"New CRS : {src.crs}")


```
